A node is simply a small, modular program that has a specific function in the grand scheme of the entire robot/system.

    i.e. A simple running program with one responsibility.

For our rover:

```
Instead of writing one giant 20,000-line program, we split it up:

LiDAR node
    |
Reads the LiDAR sensor

Camera node
    |
Processes images

Drive node
    |
Controls the motors

Localization node
    |
Figures out where the robot is

Navigation node
    |
Plans where to go

Tablet UI node
    |
Lets the user click buttons
```

Each one is its own executable.

They don't know how the others work internally.

They only communicate.

That's ROS.

Nodes communicate through three main ways: Nodes, topics, actions.

## Topics

Topic: Public channel called #weather, for instance. If someone says "It's raining" anyonein the channel can see it. nobody will reply directly / no one even knows who is listening. they just post to it, and can also read the other messages. that's a topic. the weather is the topic at hand.

Robotics ex:

LiDAR node: Reads the laser sensor. 30 times per second it publishes: Laser reading, laser reading laser reading, etc. onto /scan topic. This allows, for instance, both the mapping, navigation, and obstacle nodes to subscribe, all receiving the same data. LiDAR node doesn't care who is reading it/listening, it just posts it.

### Publisher / Subscriber

These ONLY belong to topics.

Self-explanatory. Publisher publishes, subscriber subscribes. It's just a way of referring to the process desribed above (labels):

ex.:

```
LiDAR node
      |
      | publishes
      v

    /scan
      ^
      |
Mapping node subscribes

Navigation node subscribes

Obstacle node subscribes
```

Notice the topic sits in the middle.

The nodes never talk directly.

## Services

Sometimes we do not want / need continuous updates. We just need to ask a question. 

Ex.
Tablet (UI -> human): Can we reset the odometry?

Localization node: Okay. Done.

or:

Tablet (UI -> human): What's the battery %?

Power node: 83%

This follows **REQUEST -> RESPONSE** as an architecture. Phone call, question posed, answered, hang up.

## Actions

Imagine we have something that takes time. I tell the robot: Drive 10 meters.

Such a task will not finish instantly.

Might take something like, say, 12 seconds. We likely want updates wjile it's being done, however. So:

Goal -> feedback (optional) -> Result

Say we have this scenario:

Tablet (UI -> human): Go to the kitchen.

Navigation node:

```
Accepted

Avoiding obstacle

Halfway there

2 meters left

1 meter left

Goal reached. In the kitchen.
```

Actions are for

Anything that

- takes time
- may be cancelled
- should report progress




Recap:

toic: continuous info (LiDAR, camera, gps, motor speed, imu data, temperature, joystick, etc. -> anything streaming)

service: one q, one a. reset the map pls, take a photo, change a mode, get battery, turn headlights on. quick and simple.

Action: Long task. Drive to waypoint, pick up object, explore room, dock for charging, follow a dog, etc.

```
Communication

├── Topic
│      ├── Publisher
│      └── Subscriber
│
├── Service
│      ├── Client
│      └── Server
│
└── Action
       ├── Client
       └── Server
```

Note: a given node can use a combo of all three (with varying roles as subscriber/publisher (topics), client/server (services, actions))

```
Navigation Node

✓ Subscribes to /scan        (topic)
✓ Subscribes to /odom        (topic)
✓ Publishes /cmd_vel         (topic)

✓ Offers "clear_map"         (service)

✓ Offers "navigate_to_pose"  (action)
```
